## **Importing Dataset**

```python
import kagglehub
import pandas as pd

# importing dataset
path = kagglehub.dataset_download("rounakbanik/pokemon")
print("Path to dataset files:", path)

df = pd.read_csv(f"{path}/pokemon.csv")
```

## **Data Analysis**

### **Feature & Datatype Checking**

```python
display(df.head())

print("\n--- DATA INFO ---")
df.info()
```

### **Data Description & Null Values**

```python
print("\n--- DATA DESCRIPTION ---")
display(df.describe())

print("\n--- MISSING VALUES (NULLS) ---")
print(df.isnull().sum())
```

## **Task # 1: Identifying Strong Offensive and Defensive Types**

```python
import matplotlib.pyplot as plt
import seaborn as sns

type1_stats = df.groupby('type1')[['attack', 'defense']].mean()

type1_stats = type1_stats.sort_values(by='attack', ascending=False)

print("--- AVERAGE ATTACK & DEFENSE BY PRIMARY TYPE ---")
display(type1_stats.round(2)) # 2 digits after decimal

plt.figure(figsize=(14, 7))
# use color code because default shades are too saturated
type1_stats.plot(kind='bar', figsize=(15, 6), color=['#e74c3c', '#3498db'], width=0.8)

plt.title("Average Attack and Defense by Primary Type", fontsize=16, fontweight='bold')
plt.xlabel("Primary Type (Type 1)", fontsize=12)
plt.ylabel("Average Stat Value", fontsize=12)
plt.xticks(rotation=45)
plt.legend(["Average Attack", "Average Defense"], loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

## **Task # 2: Understanding Legendary Concentration by Type**

```python
# crosstab for splitting columns (False/0 = Non-Legendary, True/1 = Legendary)
legendary_counts = pd.crosstab(df['type1'], df['is_legendary'])
legendary_counts.columns = ['Non-Legendary', 'Legendary']

legendary_counts = legendary_counts.sort_values(by='Legendary', ascending=False)

# Print Table Summary
print("--- LEGENDARY CONCENTRATION BY TYPE 1 ---")
display(legendary_counts)

# Plotting Stacked Bar Chart
sns.set_theme(style="whitegrid", rc={"grid.linestyle": "--", "grid.alpha": 0.3})
fig, ax = plt.subplots(figsize=(14, 6))

# Soft pastel colors: Greyish for Non-Legendary, Soft Gold for Legendary
legendary_counts.plot(kind='bar', stacked=True, ax=ax, color=['#B0BEC5', '#FFD700'], width=0.7, edgecolor='none')

# Styling
plt.title("Legendary Concentration by Pokémon Primary Type", fontsize=15, fontweight='bold', pad=15, color='#2C3E50')
plt.xlabel("Primary Type (Type 1)", fontsize=11, labelpad=10)
plt.ylabel("Number of Pokémon", fontsize=11, labelpad=10)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend(title="Status", labels=["Non-Legendary", "Legendary"], frameon=True, facecolor='white', edgecolor='none')

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

## **Task # 3: Finding the Most Durable and Fastest Secondaries**

```python
# Dropping rows in which  type2 have NaN's
type2_df = df.dropna(subset=['type2'])

# Custom aggregation
sec_type_stats = type2_df.groupby('type2').agg(
    Avg_HP=('hp', 'mean'),
    Max_Speed=('speed', 'max'),
    Avg_Speed=('speed', 'mean'),
    Total_Count=('name', 'count')
)

# Data Sorting
sec_type_stats = sec_type_stats.sort_values(by='Avg_Speed', ascending=False)

print("--- SECONDARY TYPE STRATEGIC METRICS ---")
display(sec_type_stats.round(2))
```

```python
sns.set_theme(style="whitegrid", rc={"grid.linestyle": "--", "grid.alpha": 0.4})

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

speed_sorted = sec_type_stats.sort_values(by='Avg_Speed', ascending=False)
sns.barplot(
    x=speed_sorted.index,
    y=speed_sorted['Avg_Speed'],
    ax=ax1,
    color='plum', # Soft Purple
    edgecolor='none'
)
ax1.set_title("Average Speed by Secondary Type ", fontsize=13, fontweight='bold', pad=10)
ax1.set_xlabel("Secondary Type (Type 2)", fontsize=11)
ax1.set_ylabel("Average Speed", fontsize=11)
ax1.tick_params(axis='x', rotation=45)

count_sorted = sec_type_stats.sort_values(by='Total_Count', ascending=False)
sns.barplot(
    x=count_sorted.index,
    y=count_sorted['Total_Count'],
    ax=ax2,
    color='lightblue', # Direct name: Soft Blue
    edgecolor='none'
)
ax2.set_title("Total Pokémon Available (Popularity)", fontsize=13, fontweight='bold', pad=10)
ax2.set_xlabel("Secondary Type (Type 2)", fontsize=11)
ax2.set_ylabel("Total Count", fontsize=11)
ax2.tick_params(axis='x', rotation=45)

sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

## **Task # 4: Custom Focus Based on Strategic Type Needs**


```python
import pandas as pd

# Custom function that checks group name directly without column dependency
def get_strategic_stat(group):
    pokemon_type = group.name

    if pokemon_type == 'fire':
        val = group['sp_attack'].mean()
        stat_name = 'Avg Special Attack'
    elif pokemon_type == 'water':
        val = group['defense'].max()
        stat_name = 'Max Defense'
    elif pokemon_type == 'electric':
        val = group['speed'].min()
        stat_name = 'Min Speed'
    else:
        val = group['attack'].median()
        stat_name = 'Median Attack'

    return pd.Series({'Strategic_Metric': stat_name, 'Value': round(val, 2)})

strategic_summary = df.groupby('type1').apply(get_strategic_stat, include_groups=False).reset_index()

print("--- STRATEGIC CUSTOM SUMMARY ---")
display(strategic_summary)
```

## **Task # 5: Evolution Across Generations**

```python
# Pivot Table
gen_type_pivot = pd.pivot_table(
    df,
    values='base_total',
    index='type1',
    columns='generation',
    aggfunc='mean'
)

# Displaying raw table for reference
print("--- EVOLUTION PIVOT TABLE (AVG BASE TOTAL) ---")
display(gen_type_pivot.round(1))

# Visualization using a Heatmap (Best for showing density and peaks)
plt.figure(figsize=(14, 9))
sns.heatmap(
    gen_type_pivot,
    annot=True,          # for numbers inside the box
    fmt=".0f",           # round off numbers
    cmap="YlGnBu",       # Direct name: Yellow to Green to Blue palette
    linewidths=0.5,
    cbar_kws={'label': 'Average Base Total Stat'}
)

plt.title("Pokémon Evolution: Average Total Stat by Type and Generation", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Generation", fontsize=12, labelpad=10)
plt.ylabel("Primary Type (Type 1)", fontsize=12, labelpad=10)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

## **Task # 6: Strategic Investment Recommendation (Critical Thinking)**

```python
# Mid-sized carry-able (Weight between 10kg and 70kg)
# dropping NaN weights first
size_filtered_df = df[df['weight_kg'].notnull()]
mid_sized_df = size_filtered_df[(size_filtered_df['weight_kg'] >= 10) & (size_filtered_df['weight_kg'] <= 70)]

# Flying is mathematically perfect counter for both.
counter_df = mid_sized_df[(mid_sized_df['type1'] == 'flying') | (mid_sized_df['type2'] == 'flying')]

# Checking for Legendaries in perfect pool
perfect_matches = counter_df[counter_df['is_legendary'] == 1]

display(perfect_matches[['name', 'type1', 'type2', 'height_m', 'weight_kg', 'is_legendary']].sort_values(by='weight_kg'))
```

```python
plot_data = perfect_matches.sort_values('weight_kg')

sns.set_theme(style="whitegrid", rc={"grid.linestyle": "--", "grid.alpha": 0.4})
plt.figure(figsize=(10, 6))

# Horizontal Bar Chart (for better readability of Names)
sns.barplot(x='weight_kg', y='name', data=plot_data, palette='pastel', edgecolor=".2")

# Red Line that limits 70kg
plt.axvline(x=70, color='crimson', linestyle='--', linewidth=2, label='Carry-able Limit (70 kg)')

plt.title(" Weights of Perfect 'Flying' Legendaries", fontsize=15, fontweight='bold', pad=15)
plt.xlabel("Weight in Kilograms (kg)", fontsize=12)
plt.ylabel("Legendary Pokémon", fontsize=12)
plt.legend(loc='lower right')

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
